### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)
r.encoding = 'utf-8'
response = r.text
print(response)
```

```python
print("Request 取回之後該怎麼取出資料，資料型態是什麼？{}".format(type(r.text)))
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ ')
soup = BeautifulSoup(response, "html5lib")
# print(type(soup))
print("處理過後會是 html 的型態")
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'

headers = {'user-agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36"}
r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
# print(r.text)
print("解決辦法:加上適當的 headers")
```
