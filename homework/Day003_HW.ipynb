### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
*  File I/O讀出來會是有許多標籤的檔案；而 xmltodict 讀出來是類似字典的資料。
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
fh = open("./data/64_72hr_CH.xml", "r")
xml = fh.read()
fh.close()
import xmltodict
d = dict(xmltodict.parse(xml))
loc = d["cwbopendata"]["dataset"]["locations"]["location"]
ans = []
for loc in loc:
    ans.append(loc['locationName'])
print("高雄市有{}處有溫度資料".format(len(ans)))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
d = dict(xmltodict.parse(xml))
loc = d["cwbopendata"]["dataset"]["locations"]["location"]
data = {}    
for i in loc:
    data[i['locationName']] = [i["weatherElement"][0]["time"][0]['dataTime'],i["weatherElement"][0]["time"][0]['elementValue']['value']]
print(data)    
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
d = dict(xmltodict.parse(xml))
loc = d["cwbopendata"]["dataset"]["locations"]["location"]
data = {}
print(loc[0]['locationName'])  # 印出鹽埕區 #下面 i 是指不同時間
# print(loc[0]['weatherElement'][0]["time"][i]['dataTime'])  # 在編號為 i 的時間點
# print(loc[0]['weatherElement'][0]["time"][i]['elementValue']["value"]) # 在編號是 i 的時間點時的溫度。
for i in range(len(loc[0]['weatherElement'])):
    data[loc[0]['weatherElement'][0]["time"][i]['dataTime']] = loc[0]['weatherElement'][0]["time"][i]['elementValue']["value"]
print(data)
```
