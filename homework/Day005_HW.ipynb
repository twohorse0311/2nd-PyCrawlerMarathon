## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
r = requests.get(" https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text
import json
data = json.loads(response)
print(data)
print("這個 API 一次會回傳{}筆資料".format(len(data)))
print("每一筆資料包含了{}".format(data[0].keys()))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print(d["title"], d["createdAt"], "留言人數",d['commentCount'], "按讚人數", d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

r1 = requests.get(" https://www.dcard.tw/_api/forums/pet/posts?popular=false") # 非熱門
response1 = r1.text
data1 = json.loads(response1)
r2 = requests.get(" https://www.dcard.tw/_api/forums/pet/posts?popular=true") # 熱門
response2 = r2.text
data2 = json.loads(response2)
pt1, pf1 =[],[] #熱門留言人數、非熱門留言人數
pt2, pf2 = [], [] # 熱門按讚人數、非熱門按讚人數
for d in data1:
    pf1.append(d['commentCount'])
    pf2.append(d['likeCount'])
for d in data2:
    pt1.append(d['commentCount'])
    pt2.append(d['likeCount'])
print("熱門文章平均留言人數：{:.2f}人".format(sum(pt1)/len(data2)))
print("非熱門文章平均留言人數：{:.2f}人".format(sum(pf1)/len(data1)))
print("熱門文章平均按讚人數：{:.2f}人".format(sum(pt2)/len(data2)))
print("非熱門文章平均按讚人數：{:.2f}人".format(sum(pf2)/len(data1)))
```
